```python
##Calculate Skill Count. T
# Then clean up the skills.
# Now we'll calculate the skill count for each data job.
# We'll reset_index to transform skills_count from a Series to a DataFrame.
# We'll prep for plotting by the data getting in descending order.
# Finally, we'll graph the top 10 skills using a bar chart.

import ast
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])

df['job_skills'] = df['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)
```

```python
df_skills = df.copy()
```

```python
df_skills = df_skills.explode('job_skills')
df_skills
```

```python
skills_counts = pd.pivot_table(df_skills, columns = ['job_skills','job_title_short'], aggfunc = 'count')
skills_counts 
```

```python
df
```

```python
df_exploded = df.explode('job_skills')
```

```python
skills_count = df_exploded.groupby(['job_title_short','job_skills']).size()
type(skills_count)
```

```python
df_skills_count = skills_count.reset_index(name = 'skill_count')
df_skills_count
df_skills_count = df_skills_count.sort_values(by = 'skill_count', ascending = False)
df_skills_count
```

```python
job_title = 'Data Analyst'
top_skills = 10

df_skills_final = df_skills_count[df_skills_count['job_title_short'] == job_title ].head(top_skills)
df_skills_final
```

```python
df_skills_final.plot(kind = 'barh', x = 'job_skills', y = 'skill_count')
```

```python
job_title = 'Data Engineer'
top_skills = 10

df_skills_final = df_skills_count[df_skills_count['job_title_short'] == job_title ].head(top_skills)
df_skills_final.plot(kind = 'barh', x = 'job_skills', y = 'skill_count')
plt.title(f"{top_skills} Top skills  for {job_title}")
plt.xlabel('Job Posting Count')
plt.legend().set_visible(False)
plt.show()
```
