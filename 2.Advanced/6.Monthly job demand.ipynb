```python
import pandas as pd 
from datasets import load_dataset
import matplotlib.pyplot as plt 

dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
df_usa = df[df['job_country'] == 'United States'].copy()
df_usa
```

```python
df_usa['job_posted_month'] = df_usa['job_posted_date'].dt.strftime('%B')
df_usa
```

```python
df_usa_pivot = df_usa.pivot_table(index = 'job_posted_month', columns = 'job_title_short', aggfunc = 'size')
df_usa_pivot
```

```python
df_usa_pivot = df_usa_pivot.reset_index()
df_usa_pivot
```

```python
#column 'month_np' is created as a datetime format of month, in order to get related to the column 'job_posted_month whi is also a datetime month format
df_usa_pivot['month_no'] = pd.to_datetime(df_usa_pivot['job_posted_month'], format = '%B').dt.month
df_usa_pivot
```

```python
df_usa_pivot.sort_values('month_no', inplace = True)
df_usa_pivot
```

```python
df_usa_pivot.drop(columns = 'month_no', inplace=True)
df_usa_pivot
```

```python
top_3 = df_usa['job_title_short'].value_counts().head(3)
top_3 = top_3.index.to_list()
top_3
```

```python
df_usa_pivot[top_3].plot(kind = 'line')
```

```python
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

df['job_posted_date'] = pd.to_datetime(df['job_posted_date']) 
```

```python
df_us = df[df['job_country'] == 'United States'].copy()
df_us
```

```python
df_us['job_posted_month'] = df_us['job_posted_date'].dt.to_period('M')
df_us
```

```python
df_us_pivot = df_us.pivot_table(index = 'job_posted_month', columns = 'job_title_short', aggfunc = 'size')
df_us_pivot
```

```python
df_us_pivot.plot(kind = 'line')
```

```python
top_3 = df_us['job_title'].value_counts().head(3).index.to_list()
top_3
```

```python
df_us_pivot = df_us_pivot[top_3]
df_us_pivot
```

```python
df_us_pivot.plot(kind='line')
plt.title('Monthly demand for 3 top jobs')
plt.xticks(rotation = 45, ha = 'right')
plt.legend()
plt.show()
```
