```python
import pandas as pd 
from datasets import load_dataset
import matplotlib.pyplot as plt 

dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
df_country_salary = df.pivot_table(values = 'salary_year_avg', index = 'job_country', columns = 'job_title_short', aggfunc = 'median')
```

```python
top_countries = df['job_country'].value_counts().head(6).index
print(top_countries)
```

```python
df_country_salary = df_country_salary.loc[top_countries]
df_country_salary
```

```python
list_jobs = ['Data Analyst', 'Data Scientist', 'Data Engineer']
df_country_salary = df_country_salary[list_jobs]
df_country_salary
```

```python
df_country_salary.plot(kind = 'bar')
plt.title('Median salaries by job title and country')
plt.xlabel('')
plt.ylabel('Job Countries')
plt.xticks(rotation = 45, ha = 'right')
plt.show()
```
