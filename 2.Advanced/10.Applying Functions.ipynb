```python
#clean all job_skills elements
import ast

def clean(skill_list):
    if pd.notna(skill_list):
        return ast.literal_eval(skill_list)

df['job_skills'] = df['job_skills'].apply(clean)
```

```python
# Importing Libraries
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt  

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
help(df.apply)
```

```python
df_salary = df.copy()
df_salary
```

```python
def projected_salary(row):
    if "Senior" in df_salary['job_title_short']:
        return 1.05 * row['salary_year_avg']
    else:
        return 1.03 * row['salary_year_avg']

df_salary['inflated_salary'] = df_salary.apply(projected_salary, axis = 1)
df_salary[['job_title_short', 'inflated_salary', 'salary_year_avg']]
```

```python
df_salary['inflated_salary'] = df_salary.apply(lambda row: 1.05 * row['salary_year_avg'] if "Senior" in row['job_title_short'] else 1.03 * row['salary_year_avg'], axis = 1)
df_salary['inflated_salary']
```
